```python
path_name = "/Users/adele/Desktop/CS325b_project/CS325B-Big-Brother-Air-Quality-Group-/datasets/"
# please update this path
npy_file_in = "ALL_EXAMPLES_for_week6_presentation_dataset.npy"
npy_array = np.load(os.path.join(path_name, npy_file_in))

npy_array.shape
col_names = "filePath webcamId webcamLat webcamLong year date hour range pm"
fmt_2 = " ".join(['%s'] * 9)
txt_file_out = "ALL_EXAMPLES_for_week6_presentation_dataset.txt"

np.savetxt(os.path.join(path_name, txt_file_out), npy_array, header = col_names, fmt = fmt_2,  comments = '')
```

```python
pm_data = pd.read_csv(os.path.join(path_name, txt_file_out), sep = " ")
print (pm_data)
list (pm_data)
```
