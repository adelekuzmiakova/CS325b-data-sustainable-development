```python
def convert_To_datetime(entry):
    res = re.match('(\d\d\d\d)(\d\d)(\d\d)', entry)
    year = int(res.group(1))
    month = int(res.group(2))
    day = int(res.group(3))
    return datetime(year, month, day)
```

```python
def convert_To_air_quality_index(pm_concentration):
    c = pm_concentration * 1.0
    c_low = None
    c_high = None
    I_low = None
    I_high = None
    if 0 < pm_concentration and pm_concentration < 12:
        return 0
    elif 12.1 < pm_concentration and pm_concentration < 33.5:
        return 1
    elif 35.5 < pm_concentration and pm_concentration < 55.4:
        return 2
    elif 55.5 < pm_concentration and pm_concentration < 150.4:
        return 3
    elif 150.5 < pm_concentration and pm_concentration < 250.4:
        return 4
    elif 250.5 < pm_concentration and pm_concentraiton < 350.4:
        return 5
    else:
        return 6
    
```
